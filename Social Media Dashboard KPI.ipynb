```python
N=48
facebook=data.loc[0].tolist()[1:]
google=data.loc[1].tolist()[1:]
linkdin=data.loc[2].tolist()[1:]
pintreast=data.loc[3].tolist()[1:]
twitter=data.loc[4].tolist()[1:]
youtube=data.loc[5].tolist()[1:]
ind=np.arange(N)
plt.figure(figsize=(15,6))
width=0.5
p1=plt.bar(ind,facebook,width)
p2=plt.bar(ind,google,width,bottom=facebook)
p3=plt.bar(ind,linkdin,width,bottom=[x+y for x,y in zip(facebook,google)])
p4=plt.bar(ind,pintreast,width,bottom=[x+y+z for x,y,z in zip(facebook,google,linkdin)])
p5=plt.bar(ind,twitter,width,bottom=[x+y+z+w for x,y,z,w in zip(facebook,google,linkdin,pintreast)]),
p6=plt.bar(ind,youtube,width,bottom=[x+y+z+w+a for x,y,z,w,a in zip(facebook,google,linkdin,pintreast,twitter)])

plt.ylabel('Visits')
plt.title('visits by Weeks')
plt.xticks(ind,data.columns[1:],rotation=90)
plt.yticks(np.arange(0, 1801, 200))
plt.legend((p1[0], p2[0],p3[0], p4[0],p5[0], p6[0]), tuple(data["Social Outlet"]))
plt.savefig("visits per week")
plt.show()
```
